```python
# Import necessary libraries
import os
import glob
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# User Input Parameters (CHANGE)
date_str = "20260210"  # YYYYMMDD format
start_time = "000000"  # HHMMSS format
end_time = "235959"  # HHMMSS format

# Optional: Set end_time_mask to limit x-axis display (set to None to disable)
# end_time_mask = "145200"  # HHMMSS format - cuts off display at this time
end_time_mask = end_time # for full-day plots

# Base path for acoustic data
base_path = "/Users/xylu/Desktop/Data/acoustic_vpp/"

# Convert strings to datetime objects for filtering
date_obj = datetime.strptime(date_str, "%Y%m%d")
start_time_obj = datetime.strptime(date_str + start_time, "%Y%m%d%H%M%S")
end_time_obj = datetime.strptime(date_str + end_time, "%Y%m%d%H%M%S")

# Convert mask time if provided
if end_time_mask is not None:
    end_time_mask_obj = datetime.strptime(date_str + end_time_mask, "%Y%m%d%H%M%S")
else:
    end_time_mask_obj = None

print(f"Date: {date_str}")
print(f"Time range: {start_time} - {end_time}")
print(f"Start: {start_time_obj.strftime('%Y-%m-%d %H:%M:%S')}")
print(f"End: {end_time_obj.strftime('%Y-%m-%d %H:%M:%S')}")
if end_time_mask_obj:
    print(f"Display mask: {end_time_mask_obj.strftime('%Y-%m-%d %H:%M:%S')}")
    
# Load CSV files for the specified date
print("\n" + "="*60)
print("Looking for CSV files...")
print("="*60)

# Direct path: files are stored in /acoustic_vpp/YYYYMMDD/
date_folder = os.path.join(base_path, date_str)

if not os.path.exists(date_folder):
    print(f"⚠ Folder not found: {date_folder}")
    filtered_df = None
else:
    print(f"✓ Found folder: {date_folder}")
    
    # Find all CSV files matching the date pattern
    csv_files = sorted(glob.glob(os.path.join(date_folder, f"{date_str}*.csv")))
    print(f"  Found {len(csv_files)} CSV file(s)")
    
    if len(csv_files) == 0:
        print(f"⚠ No CSV files found matching {date_str}*.csv")
        filtered_df = None
    else:
        # Load and combine all CSV files
        all_data = []
        
        for csv_file in csv_files:
            try:
                df = pd.read_csv(csv_file)
                # Convert time_datetime to datetime if it's a string (handles microseconds)
                if 'time_datetime' in df.columns:
                    df['time_datetime'] = pd.to_datetime(df['time_datetime'], format='ISO8601')
                all_data.append(df)
                print(f"  ✓ {os.path.basename(csv_file)}: {len(df)} samples")
            except Exception as e:
                print(f"  ✗ Error reading {os.path.basename(csv_file)}: {e}")
        
        if all_data:
            # Combine all data
            combined_df = pd.concat(all_data, ignore_index=True)
            combined_df = combined_df.sort_values('time_datetime').reset_index(drop=True)
            
            print(f"\n✓ Combined data: {len(combined_df)} total samples")
            print(f"  Time range: {combined_df['time_datetime'].min()} to {combined_df['time_datetime'].max()}")
            
            # Filter by time range
            mask = (combined_df['time_datetime'] >= start_time_obj) & (combined_df['time_datetime'] <= end_time_obj)
            filtered_df = combined_df[mask].reset_index(drop=True)
            
            print(f"\n✓ Filtered data (between {start_time_obj.strftime('%H:%M:%S')} and {end_time_obj.strftime('%H:%M:%S')}): {len(filtered_df)} samples")
            
            if len(filtered_df) > 0:
                print(f"  Vpp range: {filtered_df['vpp_volts'].min():.4f} - {filtered_df['vpp_volts'].max():.4f} V")
            else:
                print("  ⚠ No data after filtering by time range")
        else:
            print("⚠ No data loaded!")
            filtered_df = None
```

```python
# Plot machine variables in separate dual-axis figures
# Figure 1: Vpp + Beam Current
# Figure 2: Injection Rate + Injection Efficiency
# Figure 3: Charge + Injection Rep Rate
# Figure 4: Vpp + Injection Loss Rate
# Figure 5: Beam Life Loss (Current/Lifetime)
if filtered_df is not None and len(filtered_df) > 0 and machine_df_filtered is not None and len(machine_df_filtered) > 0:
    
    print(f"\n{'='*70}")
    print(f"Creating 7-panel daily plot")
    print(f"{'='*70}")
    
    # Create figure with 7 subplots in a row
    fig, (ax1, ax2, ax3, ax4, ax5, ax6, ax7) = plt.subplots(7, 1, figsize=(12, 20))
    
    # Add figure title with date
    date_formatted = datetime.strptime(date_str, "%Y%m%d").strftime("%Y-%m-%d")
    fig.suptitle(f'Acoustic, current, injection - {date_formatted}', fontsize=12, fontweight='bold', y=0.995)
    
    # ===== FIGURE 1: Vpp + Beam Current =====
    if 'A_BM_Current_mA' in machine_df_filtered.columns:
        # Get data
        mask_current = machine_df_filtered['A_BM_Current_mA'].notna()
        valid_times_current = machine_df_filtered.loc[mask_current, 'time_datetime']
        valid_current = machine_df_filtered.loc[mask_current, 'A_BM_Current_mA']
        
        # Plot Vpp on left y-axis
        color_vpp = 'tab:blue'
        ax1.set_ylabel('Bottom Acoustic Vpp (V)', color=color_vpp, fontsize=14, fontweight='bold')
        ax1.plot(filtered_df['time_datetime'], filtered_df['vpp_volts'], 
                linewidth=2.5, color=color_vpp, alpha=0.7, label='CH3 Vpp')
        ax1.tick_params(axis='y', labelcolor=color_vpp, labelsize=12)
        ax1.set_ylim(0, 18)
        ax1.grid(True, alpha=0.3)
        
        # Plot Beam Current on right y-axis
        ax1_right = ax1.twinx()
        color_current = 'darkorange'
        ax1_right.set_ylabel('Beam Current (mA)', color=color_current, fontsize=14, fontweight='bold')
        ax1_right.plot(valid_times_current, valid_current, 
                      linewidth=2.5, color=color_current, alpha=0.7, label=f'Beam Current ({len(valid_current)} pts)')
        ax1_right.tick_params(axis='y', labelcolor=color_current, labelsize=12)
        ax1_right.set_ylim(0, 1700)
        
        print(f"  ✓ Figure 1: Vpp ({len(filtered_df)} pts) + Beam Current ({len(valid_current)} pts)")
    
    # ===== FIGURE 2: Injection Rate + Injection Efficiency =====
    # Use pre-filtered arrays created immediately after data import
    if len(inj_rate_positive_values) > 0 and len(effi_nonzero_values) > 0:
        # Plot Injection Rate on left y-axis
        color_inj_rate = 'tab:green'
        ax2.set_ylabel('Injection Rate (mA/s)', color=color_inj_rate, fontsize=14, fontweight='bold')
        ax2.plot(inj_rate_positive_times, inj_rate_positive_values, 
                linewidth=0, marker='s', color=color_inj_rate, alpha=0.7, label=f'Inj Rate ({len(inj_rate_positive_values)} pts)')
        ax2.tick_params(axis='y', labelcolor=color_inj_rate, labelsize=12)
        ax2.set_ylim(0, 6)
        ax2.grid(True, alpha=0.3)
        
        # Plot Injection Efficiency on right y-axis
        ax2_right = ax2.twinx()
        color_effi = 'tab:red'
        ax2_right.set_ylabel('Injection efficiency (%)', color=color_effi, fontsize=14, fontweight='bold')
        ax2_right.plot(effi_nonzero_times, effi_nonzero_values, 
                      linewidth=0, marker='o', color=color_effi, alpha=0.7, label=f'Efficiency ({len(effi_nonzero_values)} pts)')
        ax2_right.tick_params(axis='y', labelcolor=color_effi, labelsize=12)
        ax2_right.set_ylim(0, 110)
        
        print(f"  ✓ Figure 2: Inj Rate ({len(inj_rate_positive_values)} pts) + Efficiency ({len(effi_nonzero_values)} pts)")
    
    # ===== FIGURE 3: Charge + Injection Rep Rate =====
    if 'A_Qep_BT_end_nC' in machine_df_filtered.columns and 'A_INJ_Rep_ep_Hz' in machine_df_filtered.columns:
        # Get Charge
        mask_charge = machine_df_filtered['A_Qep_BT_end_nC'].notna()
        valid_times_charge = machine_df_filtered.loc[mask_charge, 'time_datetime']
        valid_charge = machine_df_filtered.loc[mask_charge, 'A_Qep_BT_end_nC']
        
        # Get Rep Rate
        mask_rep = machine_df_filtered['A_INJ_Rep_ep_Hz'].notna()
        valid_times_rep = machine_df_filtered.loc[mask_rep, 'time_datetime']
        valid_rep = machine_df_filtered.loc[mask_rep, 'A_INJ_Rep_ep_Hz']
        
        # Plot Charge on left y-axis
        color_charge = 'tab:brown'
        ax3.set_ylabel('Charge BT end (nC)', color=color_charge, fontsize=14, fontweight='bold')
        ax3.plot(valid_times_charge, valid_charge, 
                linewidth=0, marker = "o", color=color_charge, alpha=0.7, label=f'Charge ({len(valid_charge)} pts)')
        ax3.tick_params(axis='y', labelcolor=color_charge, labelsize=12)
        ax3.set_ylim(0, 3)
        ax3.grid(True, alpha=0.3)
        
        # Plot Injection Rep Rate on right y-axis
        ax3_right = ax3.twinx()
        color_rep = 'tab:purple'
        ax3_right.set_ylabel('Injection Rep Rate (Hz)', color=color_rep, fontsize=14, fontweight='bold')
        ax3_right.plot(valid_times_rep, valid_rep, 
                      linewidth=0, marker = '*', color=color_rep, alpha=0.7, label=f'Rep Rate ({len(valid_rep)} pts)')
        ax3_right.tick_params(axis='y', labelcolor=color_rep, labelsize=12)
        ax3_right.set_ylim(0, 21)
        
        ax3.set_xlabel('Time', fontsize=14, fontweight='bold')
        print(f"  ✓ Figure 3: Charge ({len(valid_charge)} pts) + Rep Rate ({len(valid_rep)} pts)")
    
    # ===== FIGURE 4: Injection Loss Rate + Beam Life Loss Rate + Net Loss Rate =====
    # Calculate Injection Loss Rate using pre-filtered positive injection rate and non-zero efficiency
    valid_times_inj_loss = pd.Series([], dtype='datetime64[ns]')
    valid_inj_loss = np.array([])
    
    if len(effi_nonzero_values) > 0 and len(inj_rate_positive_values) > 0:
        # Create dataframe from filtered arrays (already positive injection rate and non-zero efficiency)
        df_matched = pd.DataFrame({
            'time_datetime': effi_nonzero_times,
            'efficiency': effi_nonzero_values
        })
        
        # Merge with positive injection rates to find exact timestamp matches
        df_inj = pd.DataFrame({
            'time_datetime': inj_rate_positive_times,
            'inj_rate': inj_rate_positive_values
        })
        df_matched = pd.merge(df_matched, df_inj, on='time_datetime', how='inner')
        
        if len(df_matched) > 0:
            effi_frac = df_matched['efficiency'].values / 100.0
            df_matched['inj_loss_rate'] = (1 - effi_frac) / effi_frac * df_matched['inj_rate'].values
            
            valid_loss_mask = (df_matched['inj_loss_rate'].notna())
            df_matched_valid = df_matched[valid_loss_mask].copy()
            
            valid_times_inj_loss = df_matched_valid['time_datetime']
            valid_inj_loss = df_matched_valid['inj_loss_rate'].values
    
    # Calculate Beam Life Loss Rate
    valid_times_beam_loss = pd.Series([], dtype='datetime64[ns]')
    valid_beam_loss = np.array([])
    
    if 'A_BM_Current_mA' in machine_df_filtered.columns and 'A_BM_Lifetime_min' in machine_df_filtered.columns:
        mask_current_life = (machine_df_filtered['A_BM_Current_mA'].notna()) & (machine_df_filtered['A_BM_Lifetime_min'].notna())
        mask_current_life = mask_current_life & (machine_df_filtered['A_BM_Lifetime_min'] > 0)
        
        valid_times_beam_loss = machine_df_filtered.loc[mask_current_life, 'time_datetime']
        valid_current_life = machine_df_filtered.loc[mask_current_life, 'A_BM_Current_mA']
        valid_lifetime = machine_df_filtered.loc[mask_current_life, 'A_BM_Lifetime_min']
        
        valid_beam_loss = (valid_current_life / valid_lifetime / 60.0).values
    
    # Calculate Net Loss Rate = Injection Loss Rate - Beam Life Loss Rate (only at exact matching timestamps)
    valid_times_net = pd.Series([], dtype='datetime64[ns]')
    valid_net_loss = np.array([])
    
    if len(valid_times_inj_loss) > 0 and len(valid_times_beam_loss) > 0:
        # Create DataFrames
        df_inj = pd.DataFrame({'time_datetime': valid_times_inj_loss, 'inj_loss': valid_inj_loss})
        df_beam = pd.DataFrame({'time_datetime': valid_times_beam_loss, 'beam_loss': valid_beam_loss})
        
        # Use inner merge for exact timestamp matching only
        df_net_merged = pd.merge(
            df_inj,
            df_beam,
            on='time_datetime',
            how='inner'
        )
        
        if len(df_net_merged) > 0:
            df_net_merged['net_loss'] = df_net_merged['inj_loss'] - df_net_merged['beam_loss']
            valid_times_net = df_net_merged['time_datetime']
            valid_net_loss = df_net_merged['net_loss'].values
    
    # Plot all three loss rates on the same axis
    ax4.set_ylabel('Loss Rate (mA/s)', fontsize=14, fontweight='bold')
    ax4.tick_params(axis='y', labelsize=12)
    ax4.grid(True, alpha=0.3)
    
    # Plot Injection Loss Rate
    color_inj_loss = 'tab:green'
    ax4.plot(valid_times_inj_loss, valid_inj_loss, 
            linewidth=0, marker='o', markersize=3, color=color_inj_loss, alpha=0.7, label=f'Inj Loss ({len(valid_inj_loss)} pts)')
    
    # Plot Beam Life Loss Rate
    color_beam_loss = 'tab:orange'
    ax4.plot(valid_times_beam_loss, valid_beam_loss, 
            linewidth=0, marker ="s", color=color_beam_loss, alpha=0.7, label=f'Beam Loss ({len(valid_beam_loss)} pts)')
    
    # Plot Net Loss Rate
    if len(valid_net_loss) > 0:
        ax4.plot(valid_times_net, valid_net_loss, 
            linewidth=0, marker='*', markersize=3, color='tab:red', alpha=0.7, label=f'Net Loss ({len(valid_net_loss)} pts)')
    
    # Add legend to distinguish the three loss rates
    ax4.legend(loc='upper left', fontsize=10)
    ax4.set_xlabel('Time', fontsize=14, fontweight='bold')
    ax4.set_ylim(-5, 5)
    print(f"  ✓ Figure 4: Inj Loss ({len(valid_inj_loss)} pts) + Beam Loss ({len(valid_beam_loss)} pts) + Net Loss ({len(valid_net_loss)} pts)")
    
    # ===== FIGURE 5: Beam Lifetime + Beam Current =====
    if 'A_BM_Lifetime_min' in machine_df_filtered.columns and 'A_BM_Current_mA' in machine_df_filtered.columns:
        # Get Lifetime data
        mask_lifetime = machine_df_filtered['A_BM_Lifetime_min'].notna() & (machine_df_filtered['A_BM_Lifetime_min'] > 0)
        valid_times_lifetime = machine_df_filtered.loc[mask_lifetime, 'time_datetime']
        valid_lifetime_data = machine_df_filtered.loc[mask_lifetime, 'A_BM_Lifetime_min']
        
        # Get Current data
        mask_current = machine_df_filtered['A_BM_Current_mA'].notna()
        valid_times_current_fig5 = machine_df_filtered.loc[mask_current, 'time_datetime']
        valid_current_fig5 = machine_df_filtered.loc[mask_current, 'A_BM_Current_mA']
        
        # Plot Beam Lifetime on left y-axis
        color_lifetime = 'tab:cyan'
        ax5.set_ylabel('Beam Lifetime (min)', color=color_lifetime, fontsize=14, fontweight='bold')
        ax5.plot(valid_times_lifetime, valid_lifetime_data, 
                linewidth=0, marker="o", color=color_lifetime, alpha=0.7, label=f'Lifetime ({len(valid_lifetime_data)} pts)')
        ax5.tick_params(axis='y', labelcolor=color_lifetime, labelsize=12)
        ax5.grid(True, alpha=0.3)
        ax5.set_ylim(0, 200)
        
        # Plot Beam Current on right y-axis
        ax5_right = ax5.twinx()
        color_current_fig5 = 'darkorange'
        ax5_right.set_ylabel('Beam Current (mA)', color=color_current_fig5, fontsize=14, fontweight='bold')
        ax5_right.plot(valid_times_current_fig5, valid_current_fig5, 
                      linewidth=2.5, color=color_current_fig5, alpha=0.7, label=f'Current ({len(valid_current_fig5)} pts)')
        ax5_right.tick_params(axis='y', labelcolor=color_current_fig5, labelsize=12)
        ax5_right.set_ylim(0, 1700)
        
        ax5.set_xlabel('Time', fontsize=14, fontweight='bold')
        print(f"  ✓ Figure 5: Lifetime ({len(valid_lifetime_data)} pts) + Current ({len(valid_current_fig5)} pts)")

    # ===== FIGURE 6: Total Injection Rate, Beam Life Loss, Injection Rate, Net Loss =====
    frev = 99.39e3
    valid_times_total_inj = pd.Series([], dtype='datetime64[ns]')
    valid_total_inj = np.array([])

    if 'A_INJ_Rep_ep_Hz' in machine_df_filtered.columns and 'A_Qep_BT_end_nC' in machine_df_filtered.columns:
        mask_total_inj = (machine_df_filtered['A_INJ_Rep_ep_Hz'].notna()) & (machine_df_filtered['A_Qep_BT_end_nC'].notna())
        if mask_total_inj.any():
            df_total_inj = machine_df_filtered.loc[mask_total_inj, ['time_datetime', 'A_INJ_Rep_ep_Hz', 'A_Qep_BT_end_nC']].copy()
            df_total_inj['total_inj_rate'] = df_total_inj['A_INJ_Rep_ep_Hz'].values * frev* df_total_inj['A_Qep_BT_end_nC'].values * 1E-6
            valid_times_total_inj = df_total_inj['time_datetime']
            valid_total_inj = df_total_inj['total_inj_rate'].values

    # Use pre-filtered positive injection rate from data import
    valid_times_inj_rate_fig6 = inj_rate_positive_times
    valid_inj_rate_fig6 = inj_rate_positive_values

    valid_times_net_alt = pd.Series([], dtype='datetime64[ns]')
    valid_net_alt = np.array([])
    if len(valid_times_total_inj) > 0 and len(valid_times_beam_loss) > 0 and len(valid_times_inj_rate_fig6) > 0:
        df_total = pd.DataFrame({'time_datetime': valid_times_total_inj, 'total_inj_rate': valid_total_inj})
        df_beam = pd.DataFrame({'time_datetime': valid_times_beam_loss, 'beam_loss': valid_beam_loss})
        df_inj_rate = pd.DataFrame({'time_datetime': valid_times_inj_rate_fig6, 'inj_rate': valid_inj_rate_fig6})

        df_net_alt = pd.merge(df_total, df_beam, on='time_datetime', how='inner')
        df_net_alt = pd.merge(df_net_alt, df_inj_rate, on='time_datetime', how='inner')

        if len(df_net_alt) > 0:
            df_net_alt['net_loss_alt'] = df_net_alt['total_inj_rate'] - df_net_alt['beam_loss'] - df_net_alt['inj_rate']
            # Only keep net loss for points where injection rate >= 0
            df_net_alt = df_net_alt[df_net_alt['inj_rate'] >= 0]
            valid_times_net_alt = df_net_alt['time_datetime']
            valid_net_alt = df_net_alt['net_loss_alt'].values

    ax6.set_ylabel('Rate (mA/s)', fontsize=14, fontweight='bold')
    ax6.tick_params(axis='y', labelsize=12)
    ax6.grid(True, alpha=0.3)
    ax6.plot(valid_times_total_inj, valid_total_inj,
            linewidth=0, marker='o', markersize=3, color='tab:blue', alpha=0.7, label=f'Total Inj ({len(valid_total_inj)} pts)')
    ax6.plot(valid_times_beam_loss, valid_beam_loss,
            linewidth=0, marker='s', markersize=3, color='tab:orange', alpha=0.7, label=f'Beam Loss ({len(valid_beam_loss)} pts)')
    ax6.plot(valid_times_inj_rate_fig6, valid_inj_rate_fig6,
            linewidth=0, marker='^', markersize=3, color='tab:green', alpha=0.7, label=f'Inj Rate ({len(valid_inj_rate_fig6)} pts)')
    if len(valid_net_alt) > 0:
        ax6.plot(valid_times_net_alt, valid_net_alt,
                linewidth=0, marker='*', markersize=3, color='tab:red', alpha=0.7, label=f'Net Loss ({len(valid_net_alt)} pts)')
    ax6.legend(loc='upper left', fontsize=10)
    ax6.set_xlabel('Time', fontsize=14, fontweight='bold')
    print(f"  ✓ Figure 6: Total Inj ({len(valid_total_inj)} pts) + Beam Loss ({len(valid_beam_loss)} pts) + Inj Rate ({len(valid_inj_rate_fig6)} pts) + Net Loss ({len(valid_net_alt)} pts)")

    # ===== FIGURE 7: Imported Efficiency vs Calculated Efficiency =====
    # Only plot imported efficiency where injection rate is positive AND efficiency is non-zero
    # Use the pre-filtered arrays from data import
    valid_times_effi_imported = pd.Series([], dtype='datetime64[ns]')
    valid_effi_imported = np.array([])
    
    if len(inj_rate_positive_values) > 0:
        # Get all imported efficiency values that are not null and non-zero
        mask_effi_imported = machine_df_filtered['A_INJ_Effi'].notna()
        df_all_effi_imported = pd.DataFrame({
            'time_datetime': machine_df_filtered.loc[mask_effi_imported, 'time_datetime'],
            'effi_imported': machine_df_filtered.loc[mask_effi_imported, 'A_INJ_Effi'].values
        })
        # Filter out zero efficiency values
        df_all_effi_imported = df_all_effi_imported[df_all_effi_imported['effi_imported'] != 0]
        # Merge with positive injection rate to get only overlapping timestamps
        df_inj_pos = pd.DataFrame({'time_datetime': inj_rate_positive_times})
        df_effi_merged = pd.merge(df_inj_pos, df_all_effi_imported, on='time_datetime', how='inner')
        if len(df_effi_merged) > 0:
            valid_times_effi_imported = df_effi_merged['time_datetime']
            valid_effi_imported = df_effi_merged['effi_imported'].values

    valid_times_effi_calc = pd.Series([], dtype='datetime64[ns]')
    valid_effi_calc = np.array([])
    if len(inj_rate_positive_values) > 0 and len(valid_times_total_inj) > 0:
        df_inj_rate = pd.DataFrame({'time_datetime': valid_times_inj_rate_fig6, 'inj_rate': valid_inj_rate_fig6})
        df_total = pd.DataFrame({'time_datetime': valid_times_total_inj, 'total_inj_rate': valid_total_inj})
        df_effi_calc = pd.merge(df_inj_rate, df_total, on='time_datetime', how='inner')
        if len(df_effi_calc) > 0:
            df_effi_calc['effi_calc'] = (df_effi_calc['inj_rate'] / df_effi_calc['total_inj_rate']) * 100.0
            valid_times_effi_calc = df_effi_calc['time_datetime']
            valid_effi_calc = df_effi_calc['effi_calc'].values

    ax7.set_ylabel('Injection efficiency (%)', fontsize=14, fontweight='bold')
    ax7.tick_params(axis='y', labelsize=12)
    ax7.grid(True, alpha=0.3)
    ax7.plot(valid_times_effi_imported, valid_effi_imported,
            linewidth=0, marker='o', markersize=3, color='tab:red', alpha=0.7, label=f'Imported ({len(valid_effi_imported)} pts)')
    ax7.plot(valid_times_effi_calc, valid_effi_calc,
            linewidth=0, marker='*', markersize=3, color='tab:blue', alpha=0.7, label=f'Calculated ({len(valid_effi_calc)} pts)')
    ax7.legend(loc='upper left', fontsize=10)
    ax7.set_xlabel('Time', fontsize=14, fontweight='bold')
    ax7.set_ylim(-100, 100)
    print(f"  ✓ Figure 7: Imported Efficiency ({len(valid_effi_imported)} pts) + Calculated Efficiency ({len(valid_effi_calc)} pts)")
        
    # Mark abort events on all subplots (only SBL and BeamLoss)
    if abort_today is not None and len(abort_today) > 0:
        sbl_beamloss_events = abort_today[abort_today['Category'].isin(['SBL', 'BeamLoss'])]
        if len(sbl_beamloss_events) > 0:
            print(f"\n  ✓ Marking {len(sbl_beamloss_events)} SBL/BeamLoss abort event(s) on all panels")
            for idx, row in sbl_beamloss_events.iterrows():
                abort_time = row['abort_datetime']
                # Check if abort time is within the data range
                if abort_time >= filtered_df['time_datetime'].min() and abort_time <= filtered_df['time_datetime'].max():
                    line_color = 'red'
                    text_color = 'darkred'
                    
                    # Mark on all seven axes
                    for ax in [ax1, ax2, ax3, ax4, ax5, ax6, ax7]:
                        ax.axvline(x=abort_time, color=line_color, linestyle='--', linewidth=2, alpha=0.8, zorder=10)
                        y_pos = ax.get_ylim()[1] * 0.95
                        label_text = f"{row['Category']} {abort_time.strftime('%H:%M:%S')}"
                        ax.text(abort_time, y_pos, label_text, 
                               rotation=90, verticalalignment='top', horizontalalignment='right',
                               fontsize=10, color=text_color, fontweight='bold', 
                               bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.0, edgecolor=text_color))
    
    # Format x-axes
    for ax in [ax1, ax2, ax3, ax4, ax5, ax6, ax7]:
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
        ax.tick_params(axis='x', labelsize=12)
        ax.set_xlim(start_time_obj, end_time_obj)
    
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(f"/Users/xylu/Desktop/Data/acoustic_vpp/daily_plots/{date_str}_machine_variables_panels.png", 
               dpi=150, bbox_inches='tight')
    plt.show()
    
    print(f"\n✓ Plotted 7-panel daily summary with abort events")
else:
    print("⚠ Missing required data for plots")
```

```python
# CELL 1: Import libraries, load all data, and generate merged arrays

# Self-contained setup so this cell runs without the batch-processing cell above.
import os
import re
import glob
import pandas as pd
import numpy as np
from matplotlib.colors import TwoSlopeNorm

print("="*80)
print("CORRELATION ANALYSIS - STEP 1: Data Loading and Merging")
print("="*80)

# Paths (self-contained defaults)
base_path_corr = "/Users/xylu/Desktop/Data/acoustic_vpp/"
machine_base_path_corr = "/Users/xylu/Desktop/Data/Machine_Inj_BG_byDay"

# Rebuild date list locally
all_folders_corr = [f for f in os.listdir(base_path_corr) if os.path.isdir(os.path.join(base_path_corr, f))]
date_pattern_corr = re.compile(r'^\d{8}$')
date_folders = sorted([f for f in all_folders_corr if date_pattern_corr.match(f)])

print(f"\nFound {len(date_folders)} date folders in {base_path_corr}")

# Parameters to correlate with acoustic signal
parameters_to_analyze = [
    'A_BM_Current_mA',
    'A_BM_Inj_Rate_mAps',
    'A_INJ_Effi',
    'A_INJ_Rep_ep_Hz',
    'A_Qep_BT_end_nC'
]

# Ensure column renamer exists even if previous mapping cell was not run
if 'get_friendly_name' not in globals():
    if 'pv_mapping' in globals():
        pv_dict_local = {k: v for k, v in pv_mapping}
    else:
        # Minimal fallback mapping needed for correlation cells 9/10
        pv_dict_local = {
            'BMLDCCT:CURRENT': 'A_BM_Current_mA',
            'BMLDCCT:RATE': 'A_BM_Inj_Rate_mAps',
            'CGLINJ:EFFICIENCY': 'A_INJ_Effi',
            'LIiEV:BEAM_REP:READ:KBP': 'A_INJ_Rep_ep_Hz',
            'BTpBPM:QMD11P_K_1:NC_1Hz:C': 'A_Qep_BT_end_nC',
            'BMLDCCT:LIFE': 'A_BM_Lifetime_min'
        }
    def get_friendly_name(col_name):
        return pv_dict_local.get(col_name, col_name)

# Store merged data for each date
merged_data_by_date = {}

# Process each date
print(f"\nLoading and merging data from {len(date_folders)} dates...\n")

for date_str_corr in date_folders:
    # Load acoustic data
    date_folder_corr = os.path.join(base_path_corr, date_str_corr)
    csv_files_corr = sorted(glob.glob(os.path.join(date_folder_corr, f"{date_str_corr}*.csv")))

    if len(csv_files_corr) == 0:
        continue

    # Load and combine acoustic data
    all_data_corr = []
    for csv_file in csv_files_corr:
        try:
            df_temp = pd.read_csv(csv_file)
            if 'time_datetime' in df_temp.columns:
                df_temp['time_datetime'] = pd.to_datetime(df_temp['time_datetime'], format='ISO8601')
            all_data_corr.append(df_temp)
        except Exception:
            continue

    if not all_data_corr:
        continue

    acoustic_df_corr = pd.concat(all_data_corr, ignore_index=True)
    acoustic_df_corr = acoustic_df_corr.sort_values('time_datetime').reset_index(drop=True)

    # Load machine data from new folder structure: Machine_Inj_BG_byDay/MMDD
    mmdd_corr = date_str_corr[4:8]
    machine_folder_corr = os.path.join(machine_base_path_corr, mmdd_corr)

    if not os.path.exists(machine_folder_corr):
        continue

    machine_csv_files_corr = sorted(glob.glob(os.path.join(machine_folder_corr, "*.csv")))

    if len(machine_csv_files_corr) == 0:
        continue

    # Load machine data
    machine_data_list_corr = []
    for csv_file in machine_csv_files_corr:
        try:
            df_temp = pd.read_csv(csv_file)

            # Handle both timestamp formats seen in exports
            if '#date' in df_temp.columns:
                df_temp['#date'] = pd.to_datetime(df_temp['#date'], format='ISO8601')
                df_temp.rename(columns={'#date': 'time_datetime'}, inplace=True)
            elif 'Timestamp' in df_temp.columns:
                df_temp['Timestamp'] = pd.to_datetime(df_temp['Timestamp'], format='ISO8601')
                df_temp.rename(columns={'Timestamp': 'time_datetime'}, inplace=True)
            else:
                continue

            machine_data_list_corr.append(df_temp)
        except Exception:
            continue

    if not machine_data_list_corr:
        continue

    machine_df_corr = pd.concat(machine_data_list_corr, ignore_index=True)
    machine_df_corr = machine_df_corr.sort_values('time_datetime').reset_index(drop=True)
    machine_df_corr.rename(columns=get_friendly_name, inplace=True)

    # Merge acoustic and machine data on time_datetime
    merged_df = pd.merge_asof(
        acoustic_df_corr.sort_values('time_datetime'),
        machine_df_corr.sort_values('time_datetime'),
        on='time_datetime',
        direction='nearest',
        tolerance=pd.Timedelta('2s')
    )

    # Store merged data for this date
    merged_data_by_date[date_str_corr] = merged_df

    print(f"[{date_str_corr}] Loaded and merged {len(merged_df)} data points")

# Combine all data into a single dataframe
print(f"\nCombining data from all dates...")
if len(merged_data_by_date) > 0:
    all_merged_df = pd.concat(merged_data_by_date.values(), ignore_index=True)
    print(f"Total combined data points: {len(all_merged_df)}")
    print(f"\n✓ Data loading complete! Stored {len(merged_data_by_date)} dates.")
else:
    all_merged_df = pd.DataFrame()
    print("⚠ No merged data was created. Check source folders and file formats.")
```

```python
# CELL 2: Calculate loss rate, calculate correlations, and plot

import matplotlib.pyplot as plt
from datetime import datetime

print("\n" + "="*80)
print("CORRELATION ANALYSIS - STEP 2: Calculations and Visualization")
print("="*80)

# Parameters to correlate in this step (Q_BT_end removed)
parameters_to_analyze = [
    'A_BM_Current_mA',
    'A_BM_Inj_Rate_mAps',
    'A_INJ_Effi',
    'A_INJ_Rep_ep_Hz',
    'Inj_Loss_Rate',
    'Beam_Life_Loss_Rate',
    'Net_Loss_Rate'
]

# Calculate derived parameter: Injection Loss Rate = (1-Efficiency)/Efficiency * Injection Rate
print("\nCalculating Injection Loss Rate...")

# Calculate for combined dataframe
if 'A_INJ_Effi' in all_merged_df.columns and 'A_BM_Inj_Rate_mAps' in all_merged_df.columns:
    effi_fraction = all_merged_df['A_INJ_Effi'] / 100.0
    # Set NaN for zero efficiency to avoid division by zero
    all_merged_df['Inj_Loss_Rate'] = np.where(
        effi_fraction != 0,
        (1 - effi_fraction) / effi_fraction * all_merged_df['A_BM_Inj_Rate_mAps'],
        np.nan
    )
    # Exclude negative loss rates from downstream correlation calculations
    all_merged_df.loc[all_merged_df['Inj_Loss_Rate'] < 0, 'Inj_Loss_Rate'] = np.nan
    print(f"  ✓ Calculated Inj_Loss_Rate for combined data (zero efficiency + negative values handled)")

# Calculate Beam Life Loss Rate = Current / Lifetime / 60 (mA/s)
print("Calculating Beam Life Loss Rate...")
if 'A_BM_Current_mA' in all_merged_df.columns and 'A_BM_Lifetime_min' in all_merged_df.columns:
    valid_life_mask = (
        all_merged_df['A_BM_Current_mA'].notna()
        & all_merged_df['A_BM_Lifetime_min'].notna()
        & (all_merged_df['A_BM_Lifetime_min'] > 0)
    )
    all_merged_df['Beam_Life_Loss_Rate'] = np.where(
        valid_life_mask,
        all_merged_df['A_BM_Current_mA'] / all_merged_df['A_BM_Lifetime_min'] / 60.0,
        np.nan
    )
    print("  ✓ Calculated Beam_Life_Loss_Rate for combined data")

# Calculate Net Loss Rate = Injection Loss - Beam Life Loss (with 2s tolerance)
print("Calculating Net Loss Rate with 2s tolerance...")
if 'Inj_Loss_Rate' in all_merged_df.columns and 'Beam_Life_Loss_Rate' in all_merged_df.columns:
    # Drop existing Net_Loss_Rate to avoid duplicate columns
    if 'Net_Loss_Rate' in all_merged_df.columns:
        all_merged_df = all_merged_df.drop(columns=['Net_Loss_Rate'])

    # Extract valid values with timestamps
    inj_valid = all_merged_df[['time_datetime', 'Inj_Loss_Rate']].dropna(subset=['Inj_Loss_Rate'])
    beam_valid = all_merged_df[['time_datetime', 'Beam_Life_Loss_Rate']].dropna(subset=['Beam_Life_Loss_Rate'])

    # Merge with 2s tolerance to find nearest neighbors
    matched = pd.merge_asof(
        inj_valid.sort_values('time_datetime'),
        beam_valid.sort_values('time_datetime'),
        on='time_datetime',
        direction='nearest',
        tolerance=pd.Timedelta('2s')
    )

    # Calculate Net Loss Rate for matched pairs
    matched['Net_Loss_Rate'] = matched['Inj_Loss_Rate'] - matched['Beam_Life_Loss_Rate']

    # Merge back to original dataframe
    all_merged_df = all_merged_df.merge(
        matched[['time_datetime', 'Net_Loss_Rate']],
        on='time_datetime',
        how='left'
    )

    print(f"  ✓ Calculated Net_Loss_Rate for combined data (with 2s tolerance, {matched['Net_Loss_Rate'].notna().sum():,} points)")

# Calculate for each date's merged data
for date_str, merged_df in merged_data_by_date.items():
    if 'A_INJ_Effi' in merged_df.columns and 'A_BM_Inj_Rate_mAps' in merged_df.columns:
        effi_fraction = merged_df['A_INJ_Effi'] / 100.0
        # Set NaN for zero efficiency to avoid division by zero
        merged_df['Inj_Loss_Rate'] = np.where(
            effi_fraction != 0,
            (1 - effi_fraction) / effi_fraction * merged_df['A_BM_Inj_Rate_mAps'],
            np.nan
        )
        # Exclude negative loss rates from downstream correlation calculations
        merged_df.loc[merged_df['Inj_Loss_Rate'] < 0, 'Inj_Loss_Rate'] = np.nan

    # Beam life loss rate for each date
    if 'A_BM_Current_mA' in merged_df.columns and 'A_BM_Lifetime_min' in merged_df.columns:
        valid_life_mask = (
            merged_df['A_BM_Current_mA'].notna()
            & merged_df['A_BM_Lifetime_min'].notna()
            & (merged_df['A_BM_Lifetime_min'] > 0)
        )
        merged_df['Beam_Life_Loss_Rate'] = np.where(
            valid_life_mask,
            merged_df['A_BM_Current_mA'] / merged_df['A_BM_Lifetime_min'] / 60.0,
            np.nan
        )

    # Net loss rate for each date (with 2s tolerance)
    if 'Inj_Loss_Rate' in merged_df.columns and 'Beam_Life_Loss_Rate' in merged_df.columns:
        # Drop existing Net_Loss_Rate to avoid duplicate columns
        if 'Net_Loss_Rate' in merged_df.columns:
            merged_df = merged_df.drop(columns=['Net_Loss_Rate'])

        inj_valid = merged_df[['time_datetime', 'Inj_Loss_Rate']].dropna(subset=['Inj_Loss_Rate'])
        beam_valid = merged_df[['time_datetime', 'Beam_Life_Loss_Rate']].dropna(subset=['Beam_Life_Loss_Rate'])

        if len(inj_valid) > 0 and len(beam_valid) > 0:
            matched = pd.merge_asof(
                inj_valid.sort_values('time_datetime'),
                beam_valid.sort_values('time_datetime'),
                on='time_datetime',
                direction='nearest',
                tolerance=pd.Timedelta('2s')
            )
            matched['Net_Loss_Rate'] = matched['Inj_Loss_Rate'] - matched['Beam_Life_Loss_Rate']

            merged_df = merged_df.merge(
                matched[['time_datetime', 'Net_Loss_Rate']],
                on='time_datetime',
                how='left'
            )

            # Update the stored dataframe
            merged_data_by_date[date_str] = merged_df

# Calculate daily correlations
print("\nCalculating daily correlations...")
daily_correlations = {}

for date_str, merged_df in merged_data_by_date.items():
    date_corr = {}

    for param in parameters_to_analyze:
        if param in merged_df.columns:
            valid_mask = merged_df['vpp_volts'].notna() & merged_df[param].notna()
            if valid_mask.sum() > 10:  # Need at least 10 points
                corr = merged_df.loc[valid_mask, 'vpp_volts'].corr(merged_df.loc[valid_mask, param])
                date_corr[param] = corr

    # Correlation with derived parameter (non-negative loss rate only)
    if 'Inj_Loss_Rate' in merged_df.columns:
        valid_mask = (
            merged_df['vpp_volts'].notna()
            & merged_df['Inj_Loss_Rate'].notna()
            & (merged_df['Inj_Loss_Rate'] >= 0)
        )
        if valid_mask.sum() > 10:
            corr = merged_df.loc[valid_mask, 'vpp_volts'].corr(merged_df.loc[valid_mask, 'Inj_Loss_Rate'])
            date_corr['Inj_Loss_Rate'] = corr

    daily_correlations[date_str] = date_corr

print(f"  ✓ Calculated correlations for {len(daily_correlations)} dates")

# Calculate overall correlations (all dates combined)
print("\nCalculating overall correlations...")
overall_corr = {}

for param in parameters_to_analyze:
    if param in all_merged_df.columns:
        valid_mask = all_merged_df['vpp_volts'].notna() & all_merged_df[param].notna()
        if valid_mask.sum() > 10:
            corr = all_merged_df.loc[valid_mask, 'vpp_volts'].corr(all_merged_df.loc[valid_mask, param])
            overall_corr[param] = corr

# Overall correlation with derived parameter (non-negative loss rate only)
if 'Inj_Loss_Rate' in all_merged_df.columns:
    valid_mask = (
        all_merged_df['vpp_volts'].notna()
        & all_merged_df['Inj_Loss_Rate'].notna()
        & (all_merged_df['Inj_Loss_Rate'] >= 0)
    )
    if valid_mask.sum() > 10:
        corr = all_merged_df.loc[valid_mask, 'vpp_volts'].corr(all_merged_df.loc[valid_mask, 'Inj_Loss_Rate'])
        overall_corr['Inj_Loss_Rate'] = corr

print(f"  ✓ Calculated {len(overall_corr)} overall correlations")

# Print overall correlations
print("\n" + "="*80)
print("OVERALL CORRELATIONS (All Dates Combined)")
print("="*80)
print(f"{'Parameter':<30} {'Correlation':>15} {'Strength':>15}")
print("-"*80)

# Helper function to classify correlation strength
def corr_strength(r):
    abs_r = abs(r)
    if abs_r >= 0.7:
        return "Strong"
    elif abs_r >= 0.4:
        return "Moderate"
    elif abs_r >= 0.2:
        return "Weak"
    else:
        return "Very Weak"

# Friendly names for display
param_names = {
    'A_BM_Current_mA': 'Beam Current',
    'A_BM_Inj_Rate_mAps': 'Injection Rate',
    'A_INJ_Effi': 'Injection Efficiency',
    'A_INJ_Rep_ep_Hz': 'Injection Rep Rate',
    'Inj_Loss_Rate': 'Injection Loss Rate',
    'Beam_Life_Loss_Rate': 'Beam Life Loss Rate',
    'Net_Loss_Rate': 'Net Loss Rate'
}

for param, corr in sorted(overall_corr.items(), key=lambda x: abs(x[1]), reverse=True):
    name = param_names.get(param, param)
    print(f"{name:<30} {corr:>15.4f} {corr_strength(corr):>15}")

# Create visualization
print("\n" + "="*80)
print("GENERATING VISUALIZATIONS")
print("="*80)

# Figure 1: Overall correlations bar chart + Daily correlations heatmap
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Plot 1: Overall correlations
params_display = [param_names.get(p, p) for p in overall_corr.keys()]
corr_values = list(overall_corr.values())
colors = ['red' if c < 0 else 'green' for c in corr_values]

bars = ax1.barh(params_display, corr_values, color=colors, alpha=0.7, edgecolor='black', linewidth=1.5)
ax1.axvline(x=0, color='black', linestyle='-', linewidth=1)
ax1.set_xlabel('Correlation Coefficient', fontsize=14, fontweight='bold')
ax1.set_title('Overall Correlation: Acoustic Vpp vs Machine Parameters\n(All Dates Combined)', 
              fontsize=14, fontweight='bold')
ax1.set_xlim(-1, 1)
ax1.grid(True, alpha=0.3, axis='x')
ax1.tick_params(labelsize=12)

# Add correlation values on bars
for i, (bar, val) in enumerate(zip(bars, corr_values)):
    x_pos = val + (0.05 if val > 0 else -0.05)
    ax1.text(x_pos, bar.get_y() + bar.get_height()/2, f'{val:.3f}',
            ha='left' if val > 0 else 'right', va='center', fontsize=11, fontweight='bold')

# Plot 2: Daily correlations heatmap
# Convert daily correlations to DataFrame for easier plotting
daily_corr_df = pd.DataFrame(daily_correlations).T
daily_corr_df.rename(columns=param_names, inplace=True)
daily_corr_df.index = [datetime.strptime(d, '%Y%m%d').strftime('%Y-%m-%d') for d in daily_corr_df.index]

# Plot heatmap
im = ax2.imshow(daily_corr_df.values, cmap='RdYlGn', aspect='auto', 
                norm=TwoSlopeNorm(vmin=-1, vcenter=0, vmax=1))

# Set ticks and labels
ax2.set_xticks(np.arange(len(daily_corr_df.columns)))
ax2.set_yticks(np.arange(len(daily_corr_df.index)))
ax2.set_xticklabels(daily_corr_df.columns, rotation=45, ha='right', fontsize=10)
ax2.set_yticklabels(daily_corr_df.index, fontsize=9)
ax2.set_title('Daily Correlations: Acoustic Vpp vs Machine Parameters', 
              fontsize=14, fontweight='bold')

# Add colorbar
cbar = plt.colorbar(im, ax=ax2)
cbar.set_label('Correlation Coefficient', fontsize=12, fontweight='bold')

# Add text annotations
for i in range(len(daily_corr_df.index)):
    for j in range(len(daily_corr_df.columns)):
        val = daily_corr_df.iloc[i, j]
        if pd.notna(val):
            text = ax2.text(j, i, f'{val:.2f}',
                          ha="center", va="center", color="black", fontsize=8, fontweight='bold')

plt.tight_layout()
plt.savefig('/Users/xylu/Desktop/Data/acoustic_vpp/daily_plots/correlation_analysis.png', 
           dpi=150, bbox_inches='tight')
plt.show()

print("\n✓ Correlation analysis complete!")
print(f"✓ Saved: correlation_analysis.png")
```

```python
# CELL 5: Multi-case comparison - 4 different current thresholds

from matplotlib.colors import TwoSlopeNorm

print("="*80)
print("CORRELATION COMPARISON - 4 BEAM CURRENT SCENARIOS")
print("="*80)

# Parameters for this cell (Q_BT_end removed, beam life loss added)
parameters_to_analyze = [
    'A_BM_Current_mA',
    'A_BM_Inj_Rate_mAps',
    'A_INJ_Effi',
    'A_INJ_Rep_ep_Hz',
    'Inj_Loss_Rate',
    'Beam_Life_Loss_Rate',
    'Net_Loss_Rate'
]

param_names = {
    'A_BM_Current_mA': 'Beam Current',
    'A_BM_Inj_Rate_mAps': 'Injection Rate',
    'A_INJ_Effi': 'Injection Efficiency',
    'A_INJ_Rep_ep_Hz': 'Injection Rep Rate',
    'Inj_Loss_Rate': 'Injection Loss Rate',
    'Beam_Life_Loss_Rate': 'Beam Life Loss Rate',
    'Net_Loss_Rate': 'Net Loss Rate'
}

# Ensure derived columns exist in all_merged_df
if 'Inj_Loss_Rate' not in all_merged_df.columns and 'A_INJ_Effi' in all_merged_df.columns and 'A_BM_Inj_Rate_mAps' in all_merged_df.columns:
    effi_fraction = all_merged_df['A_INJ_Effi'] / 100.0
    all_merged_df['Inj_Loss_Rate'] = np.where(
        effi_fraction != 0,
        (1 - effi_fraction) / effi_fraction * all_merged_df['A_BM_Inj_Rate_mAps'],
        np.nan
    )
    all_merged_df.loc[all_merged_df['Inj_Loss_Rate'] < 0, 'Inj_Loss_Rate'] = np.nan

if 'Beam_Life_Loss_Rate' not in all_merged_df.columns and 'A_BM_Current_mA' in all_merged_df.columns and 'A_BM_Lifetime_min' in all_merged_df.columns:
    valid_life_mask = (
        all_merged_df['A_BM_Current_mA'].notna()
        & all_merged_df['A_BM_Lifetime_min'].notna()
        & (all_merged_df['A_BM_Lifetime_min'] > 0)
    )
    all_merged_df['Beam_Life_Loss_Rate'] = np.where(
        valid_life_mask,
        all_merged_df['A_BM_Current_mA'] / all_merged_df['A_BM_Lifetime_min'] / 60.0,
        np.nan
    )

if 'Net_Loss_Rate' not in all_merged_df.columns and 'Inj_Loss_Rate' in all_merged_df.columns and 'Beam_Life_Loss_Rate' in all_merged_df.columns:
    # Drop existing Net_Loss_Rate to avoid duplicate columns
    if 'Net_Loss_Rate' in all_merged_df.columns:
        all_merged_df = all_merged_df.drop(columns=['Net_Loss_Rate'])

    # Extract valid values with timestamps
    inj_valid = all_merged_df[['time_datetime', 'Inj_Loss_Rate']].dropna(subset=['Inj_Loss_Rate'])
    beam_valid = all_merged_df[['time_datetime', 'Beam_Life_Loss_Rate']].dropna(subset=['Beam_Life_Loss_Rate'])

    # Merge with 2s tolerance to find nearest neighbors
    matched = pd.merge_asof(
        inj_valid.sort_values('time_datetime'),
        beam_valid.sort_values('time_datetime'),
        on='time_datetime',
        direction='nearest',
        tolerance=pd.Timedelta('2s')
    )

    # Calculate Net Loss Rate for matched pairs
    matched['Net_Loss_Rate'] = matched['Inj_Loss_Rate'] - matched['Beam_Life_Loss_Rate']

    # Merge back to original dataframe
    all_merged_df = all_merged_df.merge(
        matched[['time_datetime', 'Net_Loss_Rate']],
        on='time_datetime',
        how='left'
    )

# Define 4 cases to compare
cases = [
    {'lower': None, 'upper': None, 'label': 'All Data'},
    {'lower': 1000, 'upper': None, 'label': 'Current > 1000 mA'},
    {'lower': 1100, 'upper': None, 'label': 'Current > 1100 mA'},
    {'lower': 1200, 'upper': None, 'label': 'Current > 1200 mA'},
]

# Calculate correlations for each case
case_correlations = {}
case_data_counts = {}

for case in cases:
    case_label = case['label']
    lower = case['lower']
    upper = case['upper']

    # Filter data based on current range
    case_df = all_merged_df.copy()

    if lower is not None:
        case_df = case_df[case_df['A_BM_Current_mA'] >= lower]
    if upper is not None:
        case_df = case_df[case_df['A_BM_Current_mA'] <= upper]

    case_data_counts[case_label] = len(case_df)

    # Correlations for this case
    corr_dict = {}
    for param in parameters_to_analyze:
        if param in case_df.columns:
            valid_mask = case_df['vpp_volts'].notna() & case_df[param].notna()
            if valid_mask.sum() > 10:
                corr = case_df.loc[valid_mask, 'vpp_volts'].corr(case_df.loc[valid_mask, param])
                corr_dict[param] = corr

    case_correlations[case_label] = corr_dict
    print(f"[{case_label}] Data points: {len(case_df):6d}")

# Print summary table
print("\n" + "="*80)
print("SUMMARY TABLE - Correlations Across 4 Cases")
print("="*80)

all_params = [p for p in parameters_to_analyze if any(p in d for d in case_correlations.values())]

header = f"{'Parameter':<30}"
for case in cases:
    header += f" {case['label']:<18}"
print(header)
print("-" * len(header))

for param in all_params:
    row = f"{param_names.get(param, param):<30}"
    for case in cases:
        case_label = case['label']
        corr_val = case_correlations[case_label].get(param, np.nan)
        if pd.notna(corr_val):
            row += f" {corr_val:>17.4f}"
        else:
            row += f" {'N/A':>17}"
    print(row)

# Generate one correlation map (bar chart) for each case
print("\n" + "="*80)
print("GENERATING VISUALIZATIONS - One Correlation Map Per Case")
print("="*80)

fig, axes = plt.subplots(2, 2, figsize=(18, 12))
axes = axes.flatten()

for idx, case in enumerate(cases):
    case_label = case['label']
    case_corr = case_correlations[case_label]
    ax = axes[idx]

    # Prepare data for this case
    params_display = [param_names.get(p, p) for p in case_corr.keys()]
    corr_values = list(case_corr.values())
    colors = ['red' if c < 0 else 'green' for c in corr_values]

    # Create bar chart
    bars = ax.barh(params_display, corr_values, color=colors, alpha=0.7, edgecolor='black', linewidth=1.5)
    ax.axvline(x=0, color='black', linestyle='-', linewidth=1)
    ax.set_xlabel('Correlation Coefficient', fontsize=12, fontweight='bold')
    ax.set_title(f'{case_label}\n(n={case_data_counts[case_label]:,} points)', 
                fontsize=12, fontweight='bold')
    ax.set_xlim(-1, 1)
    ax.grid(True, alpha=0.3, axis='x')
    ax.tick_params(labelsize=10)

    # Add correlation values on bars
    for bar, val in zip(bars, corr_values):
        x_pos = val + (0.05 if val > 0 else -0.05)
        ax.text(x_pos, bar.get_y() + bar.get_height()/2, f'{val:.3f}', 
               ha='left' if val > 0 else 'right', va='center', 
               fontsize=9, fontweight='bold')

    print(f"  ✓ Generated correlation map for {case_label}")

fig.suptitle('Correlation Maps: Acoustic Vpp vs Machine Parameters\\nAcross Different Current Thresholds', 
            fontsize=16, fontweight='bold', y=0.995)
plt.tight_layout()
plt.savefig('/Users/xylu/Desktop/Data/acoustic_vpp/daily_plots/correlation_4case_individual_maps.png', 
           dpi=150, bbox_inches='tight')
plt.show()

print("\n✓ 4-case correlation maps complete!")
print("✓ Saved: correlation_4case_individual_maps.png")
```

```python
# CORRECTED: Properly correlate Injection Loss Rate with other parameters
# by finding nearest timestamps independently for each parameter

print("="*80)
print("CORRECTED INJECTION LOSS RATE CORRELATION ANALYSIS")
print("="*80)

# Step 1: Get timestamps where Injection Loss Rate was actually calculated
# (where both Efficiency and Injection Rate are non-null and non-zero)

print("\nStep 1: Finding timestamps with valid Injection Loss Rate...")

# Go through each day's data to find original machine timestamps
loss_rate_timestamps = []
loss_rate_values = []

for date_str in sorted(merged_data_by_date.keys()):
    df_day = merged_data_by_date[date_str]
    
    # Find rows where Injection Loss Rate exists
    valid_loss = df_day['Inj_Loss_Rate'].notna()
    
    if valid_loss.sum() > 0:
        # Get unique (timestamp, loss_rate) pairs
        loss_subset = df_day[valid_loss][['time_datetime', 'Inj_Loss_Rate']].drop_duplicates()
        loss_rate_timestamps.extend(loss_subset['time_datetime'].tolist())
        loss_rate_values.extend(loss_subset['Inj_Loss_Rate'].tolist())

# Create DataFrame of unique loss rate measurements
df_loss_rate = pd.DataFrame({
    'time_datetime': loss_rate_timestamps,
    'Inj_Loss_Rate': loss_rate_values
})

print(f"Found {len(df_loss_rate)} unique Injection Loss Rate measurements")

# Step 2: For each loss rate timestamp, find nearest Beam Current, Rep Rate, Charge
print("\nStep 2: Finding nearest machine parameters for each loss rate timestamp...")

# Collect all machine data across all days
all_machine_data = []

for date_str in sorted(merged_data_by_date.keys()):
    df_day = merged_data_by_date[date_str]
    machine_cols = ['time_datetime', 'A_BM_Current_mA', 'A_INJ_Rep_ep_Hz', 'A_Qep_BT_end_nC']
    all_machine_data.append(df_day[machine_cols])

df_all_machine = pd.concat(all_machine_data, ignore_index=True)
df_all_machine = df_all_machine.sort_values('time_datetime').reset_index(drop=True)

# For each parameter, merge with loss rate using nearest timestamp
params_to_merge = {
    'A_BM_Current_mA': 'Beam Current (mA)',
    'A_INJ_Rep_ep_Hz': 'Injection Rep Rate (Hz)',
    'A_Qep_BT_end_nC': 'Charge BT End (nC)'
}

df_corr = df_loss_rate.copy()

for param, label in params_to_merge.items():
    # Get data for this parameter (drop nulls and duplicates)
    param_data = df_all_machine[['time_datetime', param]].dropna().drop_duplicates()
    
    if len(param_data) > 0:
        # Merge using nearest timestamp within 10 seconds
        df_corr = pd.merge_asof(
            df_corr.sort_values('time_datetime'),
            param_data.sort_values('time_datetime'),
            on='time_datetime',
            direction='nearest',
            tolerance=pd.Timedelta('10s')
        )
        
        n_matched = df_corr[param].notna().sum()
        print(f"  {label}: {n_matched} matched")
    else:
        print(f"  {label}: No data available")

# Step 3: Calculate correlations
print("\n" + "="*80)
print("CORRELATION RESULTS")
print("="*80)

print(f"\n{'Parameter':<30} {'Valid Points':>15} {'Correlation':>15} {'Strength':>20}")
print("-"*80)

correlations_corrected = {}

for param, label in params_to_merge.items():
    if param in df_corr.columns:
        valid_mask = df_corr[param].notna() & df_corr['Inj_Loss_Rate'].notna()
        n_valid = valid_mask.sum()
        
        if n_valid > 10:  # Need at least 10 points for meaningful correlation
            corr = df_corr.loc[valid_mask, 'Inj_Loss_Rate'].corr(df_corr.loc[valid_mask, param])
            correlations_corrected[param] = (corr, n_valid)
            print(f"{label:<30} {n_valid:>15,} {corr:>15.4f} {corr_strength(corr):>20}")
        else:
            print(f"{label:<30} {n_valid:>15,} {'N/A':>15} {'Insufficient data':>20}")

# Step 4: Create improved scatter plots
print("\n" + "="*80) 
print("GENERATING IMPROVED SCATTER PLOTS")
print("="*80)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for idx, (param, label) in enumerate(params_to_merge.items()):
    ax = axes[idx]
    
    if param in correlations_corrected:
        corr, n_points = correlations_corrected[param]
        
        # Get valid data
        valid_mask = df_corr[param].notna() & df_corr['Inj_Loss_Rate'].notna()
        x_data = df_corr.loc[valid_mask, param]
        y_data = df_corr.loc[valid_mask, 'Inj_Loss_Rate']
        
        print(f"\n{label}: plotting {len(x_data):,} points, r={corr:.4f}")
        
        # Create scatter plot
        ax.scatter(x_data, y_data, alpha=0.6, s=40, c='darkblue', edgecolors='black', linewidth=0.5)
        
        # Add trend line
        if len(x_data) > 1:
            z = np.polyfit(x_data, y_data, 1)
            p = np.poly1d(z)
            x_trend = np.linspace(x_data.min(), x_data.max(), 100)
            ax.plot(x_trend, p(x_trend), "r--", linewidth=2.5, label='Trend line')
        
        # Labels and title
        ax.set_xlabel(label, fontsize=12, fontweight='bold')
        ax.set_ylabel('Injection Loss Rate', fontsize=12, fontweight='bold')
        ax.set_title(f'{label}\nvs Injection Loss Rate\nn={len(x_data):,} | Corr: {corr:.4f} ({corr_strength(corr)})', 
                    fontsize=11, fontweight='bold')
        ax.grid(True, alpha=0.3)
        ax.tick_params(labelsize=10)
        ax.legend(fontsize=9)
    else:
        ax.text(0.5, 0.5, f'No data for\n{label}', 
               ha='center', va='center', fontsize=12, transform=ax.transAxes)
        ax.set_xlabel(label, fontsize=12, fontweight='bold')
        ax.set_ylabel('Injection Loss Rate', fontsize=12, fontweight='bold')

plt.suptitle('Injection Loss Rate vs Machine Parameters\n(Corrected: Independent Timestamp Matching)', 
            fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('/Users/xylu/Desktop/Data/acoustic_vpp/daily_plots/injection_loss_correlation_corrected.png', 
           dpi=150, bbox_inches='tight')
plt.show()

print("\n" + "="*80)
print("✓ Corrected correlation analysis complete!")
print(f"✓ Found {len(df_loss_rate)} unique Injection Loss Rate measurements")
print("✓ Saved: injection_loss_correlation_corrected.png")
print("="*80)
```

```python
# Legacy-source comparison cell: load machine data from acoustic_vpp/YYYYMMDD/machine
# and compute correlations WITHOUT any beam-lifetime-derived metrics.

import os
import re
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from matplotlib.colors import TwoSlopeNorm

print("=" * 80)
print("CORRELATION CHECK (LEGACY MACHINE SOURCE)")
print("Machine source: /Users/xylu/Desktop/Data/acoustic_vpp/YYYYMMDD/machine")
print("Beam lifetime metrics removed from correlation calculations")
print("=" * 80)

base_path_legacy = "/Users/xylu/Desktop/Data/acoustic_vpp/"

# Rebuild date list
date_pattern = re.compile(r"^\d{8}$")
date_folders_legacy = sorted([
    d for d in os.listdir(base_path_legacy)
    if os.path.isdir(os.path.join(base_path_legacy, d)) and date_pattern.match(d)
])

# Correlation parameter list WITHOUT lifetime-based terms
parameters_to_analyze_legacy = [
    "A_BM_Current_mA",
    "A_BM_Inj_Rate_mAps",
    "A_INJ_Effi",
    "A_INJ_Rep_ep_Hz",
    "A_Qep_BT_end_nC",
    "Inj_Loss_Rate",
]

# Friendly-name fallback (no lifetime mapping needed here)
if "get_friendly_name" not in globals():
    if "pv_mapping" in globals():
        pv_dict_local = {k: v for k, v in pv_mapping}
    else:
        pv_dict_local = {
            "BMLDCCT:CURRENT": "A_BM_Current_mA",
            "BMLDCCT:RATE": "A_BM_Inj_Rate_mAps",
            "CGLINJ:EFFICIENCY": "A_INJ_Effi",
            "LIiEV:BEAM_REP:READ:KBP": "A_INJ_Rep_ep_Hz",
            "BTpBPM:QMD11P_K_1:NC_1Hz:C": "A_Qep_BT_end_nC",
        }

    def get_friendly_name(col_name):
        return pv_dict_local.get(col_name, col_name)

merged_data_by_date_legacy = {}

for date_str in date_folders_legacy:
    date_folder = os.path.join(base_path_legacy, date_str)

    # Acoustic files
    acoustic_files = sorted(glob.glob(os.path.join(date_folder, f"{date_str}*.csv")))
    if not acoustic_files:
        continue

    acoustic_chunks = []
    for f in acoustic_files:
        try:
            df = pd.read_csv(f)
            if "time_datetime" in df.columns:
                df["time_datetime"] = pd.to_datetime(df["time_datetime"], format="ISO8601")
                acoustic_chunks.append(df)
        except Exception:
            continue

    if not acoustic_chunks:
        continue

    acoustic_df = pd.concat(acoustic_chunks, ignore_index=True).sort_values("time_datetime").reset_index(drop=True)

    # Legacy machine files under YYYYMMDD/machine
    machine_folder_legacy = os.path.join(date_folder, "machine")
    if not os.path.isdir(machine_folder_legacy):
        continue

    machine_files = sorted(glob.glob(os.path.join(machine_folder_legacy, "*.csv")))
    if not machine_files:
        continue

    machine_chunks = []
    for f in machine_files:
        try:
            dfm = pd.read_csv(f)
            if "#date" in dfm.columns:
                dfm["#date"] = pd.to_datetime(dfm["#date"], format="ISO8601")
                dfm.rename(columns={"#date": "time_datetime"}, inplace=True)
            elif "Timestamp" in dfm.columns:
                dfm["Timestamp"] = pd.to_datetime(dfm["Timestamp"], format="ISO8601")
                dfm.rename(columns={"Timestamp": "time_datetime"}, inplace=True)
            else:
                continue
            machine_chunks.append(dfm)
        except Exception:
            continue

    if not machine_chunks:
        continue

    machine_df = pd.concat(machine_chunks, ignore_index=True).sort_values("time_datetime").reset_index(drop=True)
    machine_df.rename(columns=get_friendly_name, inplace=True)

    merged_df = pd.merge_asof(
        acoustic_df.sort_values("time_datetime"),
        machine_df.sort_values("time_datetime"),
        on="time_datetime",
        direction="nearest",
        tolerance=pd.Timedelta("2s"),
    )

    # Injection loss only (no lifetime-based calculations)
    if "A_INJ_Effi" in merged_df.columns and "A_BM_Inj_Rate_mAps" in merged_df.columns:
        effi_fraction = merged_df["A_INJ_Effi"] / 100.0
        merged_df["Inj_Loss_Rate"] = np.where(
            effi_fraction != 0,
            (1 - effi_fraction) / effi_fraction * merged_df["A_BM_Inj_Rate_mAps"],
            np.nan,
        )
        merged_df.loc[merged_df["Inj_Loss_Rate"] < 0, "Inj_Loss_Rate"] = np.nan

    merged_data_by_date_legacy[date_str] = merged_df

if not merged_data_by_date_legacy:
    print("No merged data found from legacy machine source.")
else:
    all_merged_df_legacy = pd.concat(merged_data_by_date_legacy.values(), ignore_index=True)
    print(f"Merged dates: {len(merged_data_by_date_legacy)}")
    print(f"Total merged points: {len(all_merged_df_legacy)}")

    # Daily correlations
    daily_correlations_legacy = {}
    for date_str, df in merged_data_by_date_legacy.items():
        date_corr = {}
        for param in parameters_to_analyze_legacy:
            if param in df.columns:
                valid = df["vpp_volts"].notna() & df[param].notna()
                if valid.sum() > 10:
                    date_corr[param] = df.loc[valid, "vpp_volts"].corr(df.loc[valid, param])
        daily_correlations_legacy[date_str] = date_corr

    # Overall correlations
    overall_corr_legacy = {}
    for param in parameters_to_analyze_legacy:
        if param in all_merged_df_legacy.columns:
            valid = all_merged_df_legacy["vpp_volts"].notna() & all_merged_df_legacy[param].notna()
            if valid.sum() > 10:
                overall_corr_legacy[param] = all_merged_df_legacy.loc[valid, "vpp_volts"].corr(
                    all_merged_df_legacy.loc[valid, param]
                )

    print("\n" + "=" * 80)
    print("OVERALL CORRELATIONS (LEGACY MACHINE SOURCE)")
    print("=" * 80)

    def corr_strength(r):
        ar = abs(r)
        if ar >= 0.7:
            return "Strong"
        if ar >= 0.4:
            return "Moderate"
        if ar >= 0.2:
            return "Weak"
        return "Very Weak"

    param_names_legacy = {
        "A_BM_Current_mA": "Beam Current",
        "A_BM_Inj_Rate_mAps": "Injection Rate",
        "A_INJ_Effi": "Injection Efficiency",
        "A_INJ_Rep_ep_Hz": "Injection Rep Rate",
        "A_Qep_BT_end_nC": "Q at BT End",
        "Inj_Loss_Rate": "Injection Loss Rate",
    }

    print(f"{'Parameter':<30} {'Correlation':>15} {'Strength':>15}")
    print("-" * 80)
    for param, corr in sorted(overall_corr_legacy.items(), key=lambda x: abs(x[1]), reverse=True):
        print(f"{param_names_legacy.get(param, param):<30} {corr:>15.4f} {corr_strength(corr):>15}")

    # Plot: bar + heatmap, matching the style of your existing correlation output
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

    display_names = [param_names_legacy.get(p, p) for p in overall_corr_legacy.keys()]
    corr_values = list(overall_corr_legacy.values())
    colors = ["red" if c < 0 else "green" for c in corr_values]

    bars = ax1.barh(display_names, corr_values, color=colors, alpha=0.7, edgecolor="black", linewidth=1.5)
    ax1.axvline(x=0, color="black", linestyle="-", linewidth=1)
    ax1.set_xlabel("Correlation Coefficient", fontsize=14, fontweight="bold")
    ax1.set_title("Overall Correlation: Acoustic Vpp vs Machine Parameters (Legacy Source)", fontsize=14, fontweight="bold")
    ax1.set_xlim(-1, 1)
    ax1.grid(True, alpha=0.3, axis="x")

    for bar, val in zip(bars, corr_values):
        x_pos = val + (0.05 if val > 0 else -0.05)
        ax1.text(
            x_pos,
            bar.get_y() + bar.get_height() / 2,
            f"{val:.3f}",
            ha="left" if val > 0 else "right",
            va="center",
            fontsize=11,
            fontweight="bold",
        )

    daily_corr_df_legacy = pd.DataFrame(daily_correlations_legacy).T
    if not daily_corr_df_legacy.empty:
        daily_corr_df_legacy.rename(columns=param_names_legacy, inplace=True)
        daily_corr_df_legacy.index = [datetime.strptime(d, "%Y%m%d").strftime("%Y-%m-%d") for d in daily_corr_df_legacy.index]

        im = ax2.imshow(
            daily_corr_df_legacy.values,
            cmap="RdYlGn",
            aspect="auto",
            norm=TwoSlopeNorm(vmin=-1, vcenter=0, vmax=1),
        )
        ax2.set_xticks(np.arange(len(daily_corr_df_legacy.columns)))
        ax2.set_yticks(np.arange(len(daily_corr_df_legacy.index)))
        ax2.set_xticklabels(daily_corr_df_legacy.columns, rotation=45, ha="right", fontsize=10)
        ax2.set_yticklabels(daily_corr_df_legacy.index, fontsize=9)
        ax2.set_title("Daily Correlations: Acoustic Vpp vs Machine Parameters (Legacy Source)", fontsize=14, fontweight="bold")

        cbar = plt.colorbar(im, ax=ax2)
        cbar.set_label("Correlation Coefficient", fontsize=12, fontweight="bold")

        for i in range(len(daily_corr_df_legacy.index)):
            for j in range(len(daily_corr_df_legacy.columns)):
                val = daily_corr_df_legacy.iloc[i, j]
                if pd.notna(val):
                    ax2.text(j, i, f"{val:.2f}", ha="center", va="center", color="black", fontsize=8, fontweight="bold")
    else:
        ax2.axis("off")
        ax2.text(0.5, 0.5, "No daily correlation matrix to display", ha="center", va="center", fontsize=12)

    plt.tight_layout()
    plt.show()

    print("\nDone: legacy-source correlation check complete (no lifetime metrics).")
```
