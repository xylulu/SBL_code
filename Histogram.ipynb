```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator

# Exact LER beam current values [mA] from your dataset
data = [
    1489, 1421, 1546, 1362, 1496, 1098, 1495, 1447, 1297, 1410,
    1446, 1394, 1346, 1396, 1396, 1097, 1396, 1397, 1338, 1396,
    1298, 1197, 998, 299
]

# Define bins: 100 mA wide, starting from 0 and covering the data range
bins = np.arange(0, 1700, 100)

# Create the plot
plt.figure(figsize=(6, 4))
plt.hist(data, bins=bins, edgecolor='black', color='skyblue', alpha=0.8)

# Formatting
plt.title('Number of beam loss/SBL events vs. LER current', fontsize=14)
plt.xlabel('LER Beam Current (mA)', fontsize=14)
plt.ylabel('Number of Events', fontsize=14)
plt.xticks(bins, rotation=45, fontsize =12)
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
plt.tick_params(axis='y', labelsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout and save/show
plt.tight_layout()
# plt.savefig('sbl_histogram_correct_24.png')
plt.show()
```

```python
# Numeric Diamond abort signal values (9 total)
diamond_numeric = [2224.5, 16.4, 774.8, 41.9, 22.2, 139.9, 13.2, 99.6, 981.6]
# Number of N/A entries (15 total)
na_count = 15

# 1. Define numeric bins (100 units wide)
bins = np.arange(0, 2400, 100)
counts, bin_edges = np.histogram(diamond_numeric, bins=bins)

# 2. Prepare labels and values for a combined bar chart
labels = []
values = []

# Add numeric bins that have data
for i in range(len(counts)):
    if counts[i] > 0:
        labels.append(f"{int(bin_edges[i])}-{int(bin_edges[i+1])}")
        values.append(counts[i])

# 3. Add the N/A column on the right
labels.append("N/A")
values.append(na_count)

# 4. Create the plot
plt.figure(figsize=(6, 4))
bars = plt.bar(labels, values, color='lightgreen', edgecolor='black', alpha=0.8)

# Add numeric count labels on top of each bar for clarity
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.2, int(yval), ha='center', va='bottom', fontweight='bold')

# Formatting
plt.title('Number of beam loss/SBL events vs. Diamond Abort Dose', fontsize=14)
plt.xlabel('Diamond Abort Dose (mRad)', fontsize=12)
plt.ylabel('Number of Events', fontsize=12)
plt.xticks(rotation=45, fontsize=12)
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
plt.tick_params(axis='y', labelsize=12)
plt.ylim(0, max(values) + 2)  # Add some space above the tallest bar
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
